```python
import pandas_datareader.data as web
import datetime
import numpy as np
import pandas as pd
start = datetime.datetime(2016, 3, 1)
end = datetime.datetime(2021, 5, 28)
skhynix = web.DataReader("000660.KS", "yahoo", start, end)
skhynix_1 = skhynix.to_numpy()
skhynix_2 = skhynix_1.tolist()
skhynix
```

```python
skhynix_9일_최저시가 =[]
skhynix_9일_최고시가 =[]
skhynix_전환선 =[]
for i in range(9,len(skhynix_2)):
    sum=[]
    for j in range(i,i+9):
        sum.append(skhynix_2[j-9][2])
    skhynix_9일_최저시가.append(min(sum))
    skhynix_9일_최고시가.append(max(sum))
    skhynix_전환선.append((max(sum)+min(sum))//2)
for i in range(9):
    skhynix_전환선.insert(0,0)
skhynix['change']=skhynix_전환선


skhynix

```

```python
skhynix_26일_최저시가 =[]
skhynix_26일_최고시가 =[]
skhynix_기준선 =[]
for i in range(26,len(skhynix_2)):
    sum=[]
    for j in range(i,i+26):
        sum.append(skhynix_2[j-26][2])
    skhynix_26일_최저시가.append(min(sum))
    skhynix_26일_최고시가.append(max(sum))
    skhynix_기준선.append((max(sum)+min(sum))//2)

for i in range(26):
    skhynix_기준선.insert(0,0)
skhynix['standard']=skhynix_기준선
skhynix
```

```python
skhynix_52일_최저시가 =[]
skhynix_52일_최고시가 =[]
skhynix_선행스팬2 =[]
for i in range(52,len(skhynix_2)):
    sum=[]
    for j in range(i,i+52):
        sum.append(skhynix_2[j-52][2])
    skhynix_52일_최저시가.append(min(sum))
    skhynix_52일_최고시가.append(max(sum))
    skhynix_선행스팬2.append((max(sum)+min(sum))//2)
    
print(len(skhynix_선행스팬2))

for i in range(52):
    skhynix_선행스팬2.insert(0,0)
    
skhynix['leading span2']=skhynix_선행스팬2
skhynix
```

```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import mpl_finance
import matplotlib.ticker as ticker

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)

day_list = []
name_list = []
t=0
for i, day in enumerate(skhynix.index):
    if day.dayofweek == 0:
        t+=1
        if t%40==0:
            day_list.append(i)
            name_list.append(day.strftime('%Y-%m-%d'))

ax.xaxis.set_major_locator(ticker.FixedLocator(day_list))
ax.xaxis.set_major_formatter(ticker.FixedFormatter(name_list))

mpl_finance.candlestick2_ohlc(ax, skhynix['Open'], skhynix['High'], skhynix['Low'], skhynix['Close'], width=0.5, colorup='r', colordown='b')
plt.grid()
plt.show()
```
