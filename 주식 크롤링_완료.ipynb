```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

LG1_url = 'https://api.finance.naver.com/siseJson.naver?symbol=066570&requestType=1&startTime=20200321&endTime=20210526&timeframe=day'
LG2_url = 'https://api.finance.naver.com/siseJson.naver?symbol=066570&requestType=1&startTime=20190721&endTime=20200323&timeframe=day'
LG1 = requests.get(LG1_url)
LG2 = requests.get(LG2_url)
LG_data1 = BeautifulSoup(LG1.text, 'lxml')
LG_data2 = BeautifulSoup(LG2.text, 'lxml')
#print (LG_data1)
#print(LG_data2)


LGLG1 = LG_data1.select_one("html > body").string
LGLG1= list(LGLG1.split(']'))
#print(LGLG1)
LGLG2 = LG_data2.select_one("html > body").string
LGLG2= list(LGLG2.split(']'))
#print(LGLG2)

LGLGLG1=[]
LGLGLG2 =[]
for i in LGLG1:
    LGLGLG1.append(i.split(',')[1:])
for i in LGLG2:
    LGLGLG2.append(i.split(',')[1:])
LGLGLG1=LGLGLG1[1:-2]
LGLGLG2=LGLGLG2[1:-2]
for i in range(len(LGLGLG1)):
    LGLGLG1[i][0]=LGLGLG1[i][0][6:-1]
for i in range(len(LGLGLG2)):
    LGLGLG2[i][0]=LGLGLG2[i][0][6:-1]
    
LGLGLG1[0][0]=LGLGLG1[0][0][5:]
LGLGLG2[0][0]=LGLGLG2[0][0][5:]

for i in range(len(LGLGLG1)):
    for j in range(len(LGLGLG1[i])):
        if j==6:
            LGLGLG1[i][j]=float(LGLGLG1[i][j])
        else:
            LGLGLG1[i][j]=int(LGLGLG1[i][j])
for i in range(len(LGLGLG2)):
    for j in range(len(LGLGLG2[i])):
        if j==6:
            LGLGLG2[i][j]=float(LGLGLG2[i][j])
        else:
            LGLGLG2[i][j]=int(LGLGLG2[i][j])
LG_final = LGLGLG2 + LGLGLG1
LG_final
print(len(LG_final))

```

```python
LG_9일_최저시가 =[]
LG_9일_최고시가 =[]
LG_전환선 =[]
for i in range(9,len(LG_final)):
    sum=[]
    for j in range(i,i+9):
        sum.append(LG_final[j-9][1])
    #print(sum)
    LG_9일_최저시가.append(min(sum))
    LG_9일_최고시가.append(max(sum))
    LG_전환선.append((max(sum)+min(sum))//2)
#print(LG_9일_최저시가, end = '\n\n\n\n\n\n')
#print(LG_9일_최고시가, end = '\n\n\n\n\n\n')
print(LG_전환선)

#print(len(LG_9일_최저시가))
#print(len(LG_9일_최고시가))
print(len(LG_전환선))
```
